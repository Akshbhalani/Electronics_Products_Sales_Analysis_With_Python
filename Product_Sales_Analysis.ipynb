```python
dataframe = df.drop(index = 16564)
```

```python
# delete the rows where rating has nan


dataframe['rating'].dropna(inplace = True)
```

```python
# changing datatype of the columns

dataframe['rating'] = dataframe['rating'].astype(int)
dataframe['user_id'] = dataframe['user_id'].astype(int)
dataframe['category'] = dataframe['category'].astype(str)
dataframe['brand'] = dataframe['brand'].astype(str)
```

```python
print(dataframe.info())
```

```python
print(dataframe)
```

```python
sns.countplot(x='rating', data=dataframe)
```

```python
# adding year column using DatetimeIndex in dataframe

dataframe['year'] = pd.DatetimeIndex(dataframe['timestamp']).year
print(dataframe)
```

```python
# counting the rating groupby year.

# INSIGHT : 2007 is the best year for selling

dataframe.groupby('year')['rating'].count().plot(kind='bar')
```

```python
# adding month column using DatetimeIndex in dataframe

dataframe['month'] = pd.DatetimeIndex(dataframe['timestamp']).month

print(dataframe)
```

```python
# INSIGHT : December and January months are the best months for selling products.

dataframe.groupby('month')['rating'].count().plot(kind = 'bar')
```

```python
dataframe_2006 = dataframe[dataframe['year'] == 2006]
```

```python
print(dataframe_2006)
```

```python
# INSIGHT : in 2006 'Kodak', 'Canon' and 'Linksys' are the favorite brands among all brands

dataframe_2006.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind = 'bar')
```

```python
print(dataframe_2006.groupby('category')['rating'].count().sort_values(ascending = False))
```

```python
# INSIGHT: 'Camera & Photo' could be the highest sold product in 2006

category_data_2006 = dataframe_2006.groupby('category')['rating'].count().sort_values(ascending=True).plot(kind = 'bar')
```
